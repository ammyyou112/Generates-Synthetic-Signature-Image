```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from faker import Faker
from PIL import Image, ImageDraw, ImageFont
import random
import os
```

```python
fake = Faker('ar_SA')
```

```python
def generate_signature(Name):
    img = Image.new('RGB', (100, 50), color=(0, 0, 0))
    draw = ImageDraw.Draw(img)
    text = f"{Name}"
    font = ImageFont.truetype("Times New Roman.ttf", 14)
    x = 10
    y = (img.height - 50) // 2
    draw.text((x, y), text, font=font, fill=(255, 255, 255))
    return img

data = []
for _ in range(100):
    Name = fake.name()
    FatherName = fake.name()
    ID = fake.iana_id()
    Address = fake.address()
    Signature = generate_signature(Name)
    Signature.save(f"{Name}.jpg") 
    data.append([Name, FatherName, ID, Address, f"{Name}.jpg"])  
```

```python
df = pd.DataFrame(data, columns=['Name', 'FatherName', 'ID', 'Address', 'Signature'])sss
df['Signature'] = df['Signature'].apply(lambda x: np.array(Image.open(x)))

for index, row in df.iterrows():
    plt.imshow(row['Signature'])
    plt.show()
```
